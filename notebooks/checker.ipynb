```python
import json
from datasets import load_from_disk
```

```python
default_dataset_path = '/opt/ml/input/data/train_dataset'
datasets = load_from_disk(default_dataset_path)
```

```python
val_data = datasets['validation']
val_data
```

```python
sample_idx = 0
print(val_data[sample_idx]['id'])
print(val_data[sample_idx]['document_id'])
print(val_data[sample_idx]['question'])
print(val_data[sample_idx]['context'])
print(val_data[sample_idx]['answers'])
```

```python
answer_dict = {sample['id']: {'question': sample['question'], 'context': sample['context'],'answers': sample['answers']['text']} for sample in val_data}
```

```python
answer_dict['mrc-0-003264']
```

```python
# answer_or_reference = ''
target = '/opt/ml/ckpts/2021-04-29-15:33:06/predictions_fixed_v2.json'
with open(target) as json_data:
    my_predictions = json.load(json_data)
```

```python
diff = {}

for p, a in my_predictions.items():
    if a not in answer_dict[p]['answers']:
        print(f'{p} Answer: {answer_dict[p]["answers"]}')
        print(f'My prediction: {a}\n')

        diff[p] = a
    
```

```python
print(f'Number of wrong answers = {len(diff)}')
```

```python
import random
from pprint import pprint

num_samples = 10
check_indices = random.sample(diff.keys(), num_samples)

for i in check_indices:
    print('{} {} {}'.format('*'*20, i, '*'*20))
    print(f"[Question] {answer_dict[i]['question']}")
    print(f'[Context]')
    pprint(answer_dict[i]['context'], compact=True)
    print(f"[Answer] {answer_dict[i]['answers']}")
    print(f'[Prediction] {diff[i]}')
    print()
```
